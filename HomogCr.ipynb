```python
import pandas as pd, scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
```

```python
data =  pd.ExcelFile("/Users/uliakaliberda/Downloads/LabsMS/Lab5.xls")
df = data.parse("1")
first = pd.DataFrame()
first["A"] = df["X1"]
first["B"] = df["X2"]
first["C"] = first["A"] - first["B"]
first[:10]
```

Criteria V-M

```python
first["D"] = sorted(abs(first["C"]))
first
```

```python
signs = []
for elemd in first["D"]:
    for i in range(len(first["C"])):
        if abs(first["C"][i])==elemd:
            if (first["C"][i])>0:
                    signs.append(1)
            else:
                    signs.append(-1)
first["E"] = signs
```

```python
U = 0; #ranks
for i in range(len(first["E"])):
    if first["E"][i]>0:
        U+=i;
U
```

```python
M = 51*50/2
D = M*51/12
D
```

```python
T = (U-M)*D**(-0.5)
T
```

```python
F = (1/2*np.pi)*(np.exp(-0.5))
F
```

```python
if F>T:
    print("H0")
else:
    print("H1")
```
